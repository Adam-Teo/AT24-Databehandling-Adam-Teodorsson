```python
import pandas as pd 
import numpy as np 

df1 = pd.DataFrame(np.arange(16).reshape(4,4), columns=list("ABCD"))
df2 = pd.DataFrame(np.zeros((3,4)), columns=list("ABCD"))
df3 = pd.DataFrame(np.ones((4,3)), index=[1,5,10,15])
df1
```

#### Concat
. Don't modify, instead return
. Combinera med LiCo if you want to do multiple concat
. Outer Join (Union) : Joins everything
. Inner Join (Intersection) : Only overlapping element ex only columns that contains data
. Concat joins vertical, concat axis = "columns" joins horizontal 
. Join is outer by default so if values are missing are set to NaN, axis = 1 (x,y)
. if floats exist the collumn then all elments become floats, or if there are NaNs
. but different columns can have different type 
. Join inner only adds ther row that exists in all 
. Append : Depricated aka does no longer exist



```python
# same column names, just add the second df rows to the first
pd.concat([df1,df2]) # by default axis = 0 or "rows"
```

```python
# concats on the index 
# NOTE: NaNs due to df2 shape is (3,4) while df1 shape is (4,4)
pd.concat([df1,df2], axis = "columns")
```

```python
# NOTE: index 1 is shared while others are not 
pd.concat([df1,df3], axis = 1)  # join="outer by default"
```

```python
# inner join - set logic intersection "A and B"
pd.concat([df1,df3], axis = 1, join="inner") 
```

#### Merge
. Union, ignores incomplete and merges on indexes, how == inner
. indicator will show what happens
> pd.merege(left, right, on="key", indicator=True)

. Union, fills incomplete with NaN and merges on indexes 
> pd.merege(left, right, on="key", how="outer", indicator=True)

. Multiple keys 
. How tells pd which values to prioriteis
> left.merge(right, on=["key1", "key2"], how="lef", indicator=True)

```python
# merge
left = pd.DataFrame({"key": ["K0", "K0", "K1", "K2"],
                     "A": ["A0", "A1", "A2", "A3"],
                     "B": ["B0", "B1", "B2", "B3"],
                     })
right = pd.DataFrame({"key": ["K0", "K1", "K2", "K3"],
                     "C": ["C0", "C1", "C2", "C3"],
                      "D": ["D0", "D1", "D2", "D3"]})
print(left)
print(right)
```

```python
# note on column must be found in both left and right
pd.merge(left, right, on="key", indicator=True) # how= "inner" default
```

```python
# note that for key K3 there is only the right one and there is NaN in col A,B because 
# outer join is a union between left and right 
left.merge(right, on=["key"], how="outer", indicator=True)
```

```python
left.merge(right, on="key", how="left", indicator=True)
```

```python
left = pd.DataFrame(
    {
        "key1": ["K0", "K0", "K1", "K2"],
        "key2": ["K0", "K1", "K0", "K1"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)

right = pd.DataFrame(
    {
        "key1": ["K0", "K1", "K1", "K2"],
        "key2": ["K0", "K0", "K0", "K0"],
        "A": ["A0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)

print(left)

print(right)

# A_x is from left, A_y is from right, since A exist in both dataframes
# only keys present in left
left.merge(right, on=["key1", "key2"], how="left", indicator=True)
```

```python
# keys in right
left.merge(right, on=["key1", "key2"], how="right", indicator=True, suffixes=["_left","_right"])
```

#### Join
. Join works in a similar way to merge, infact join uses merge
> left.join(right)
> right.join(left)

```python
left = pd.DataFrame(
    {"A": ["A0", "A1", "A2"], "B": ["B0", "B1", "B2"]}, index=["K0", "K1", "K2"]
)


right = pd.DataFrame(
    {"C": ["C0", "C2", "C3"], "D": ["D0", "D2", "D3"]}, index=["K0", "K2", "K3"]
)
left, right
```

```python
left.join(right) # joins on left
```

```python
right.join(left) # joins on right
```

```python
left.join(right, how = "right")
```
