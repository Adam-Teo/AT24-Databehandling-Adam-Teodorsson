*h2) Size of Institution vs Graduation Rate*\
I want to check how the size of the educational institution affects the\
graduation rate, does smaller institutions have a higher or lower rate\
of graduates.\
\
I'm going take the liberty of assuming that the number of enrolled\
students corresponds to the size of the school and use a scatter graph\
to plot each school and see how they compare.


```python

# h2) line graph

# There is one Grad.Rate that is at 120%, I assume that this is a mistake
# and have chosen to adjust it to 100%, technically it would probabbly 
# be more accurat to simply remove the school from the comparison 
college["Grad.Rate"] = college["Grad.Rate"].apply(lambda val: 100 if val > 100 else val)

college.sort_values("Enroll", inplace=True)
fig = px.scatter(x=college["Enroll"], y=college["Grad.Rate"], )
fig.update_layout(title='h2) Enrollment vs Graduation Rate',
                   xaxis_title='Enrollment',
                   yaxis_title='Graduation Rate (%)')
fig.show()
```

*h2) Conclusion-Scatter*\
I would say, looking att this graph, that the bigger schools from 4-6k enrollment\
don't so much have a higher graduation rate as they have a more average rate at around 75%.\
It's a lot harder to determine the average graduation rate for the schools with around\
a 1000 enrollment as there are so many points, so I'm going to plot a line graph.

```python
# h2) Line Graph


import numpy as np
college.sort_index(inplace=True)

# Creates bins for the lines because with a 777 schools the line 
# would otherwise be pretty hard to read.
college["enroll_bin"] = pd.cut( college["Enroll"], np.linspace(0,6400,16), labels=list(range(0,15)) )

# I use enroll_bin to callcualte the Grad.Rate mean for each bin
y = college.groupby("enroll_bin")["Grad.Rate"].mean()

# Because some bins end up empty I fills in the missing values 
# with the average of the value before and after
y = (y.bfill() + y.ffill())/2


fig = px.line( x=np.linspace(0,6400,15), y=y ) 

fig.update_layout(title='h2) Enrollment vs Graduation Rate',
                   xaxis_title='Enrollment',
                   yaxis_title='Graduation Rate (%)')
fig.show()
```

*h2) Conclusion-Line*\
This graph makes things a little more clear and we can see that on average\
the smallest schools (1-1000) and the largest schools(5500-6500) seem to have\
around the same graduation rate, averaging around 67%.\
The schools with around 5k enrollments seem to be doing really well but\
when you compare this data with the scatter plot it looks like the peak\
at 87% followed by the drop to 65% is based on the performance of\
only two schools so I would argue that due to this super low sample rate\
this part of the graph is pretty unreliable, however if we had data from\
spanning multiple years then this would have made it more reliable.\
\
The midsized schools with around 2-4k appears to be doing the worst and\
there is even a pretty sever dip down to 50% at around the 2.7k mark.\
\
I wouldn't want to draw any hard conclusions from this data alone, I would\
at least like to see how the graduation rate changes over the course\
of say a decade.\
But it does seem that on average midsized schools do a bit worse than\
the smaller and larger schools when it comes to graduation rate.

*h3) Past Performance vs Graduation Rates*\
Continuing the graduation rate examination I wanted to see if\
past performance of students was good indicator for chance of\
graduating.\
I was planning to use a bar plot, but I decided on a histogram\ 
instead as it has built in bin functionality.

```python
# h3)

# Removing the top 10 percentile from the 25 percentile, turning
# them in to two distinct layers without overlap.
college["Top11-25perc"] = college["Top25perc"]-college["Top10perc"]

# I wanted to fill out the graph with the student that 
# did not graduate from the top 25 percentile 
college["Top26-100perc"] = 100-college["Top25perc"]

fig = px.histogram(college, nbins=40, x="Grad.Rate", y=["Top10perc", "Top11-25perc", "Top26-100perc"], histfunc="avg" )
fig.update_layout( title='h3) Gradtauion Rate vs High School Performance',
                   xaxis_title='Graduation Rate',
                   yaxis_title='High School Performance')
fig.show()
```

*h3) Outcome*\
The outcome seem to reflect what one might assume, that the schools with a\
higher number of students from that graduate top 25% of their high school class\ 
also have a higher degree of students that manage to graduate\
What I find interesting however is the 10 to 14 graduation range has a very high\
percentage of students from the top 25% but if we explore the data using\
college.sort_values("Grad.Rate").head()\
we can see that this bin contains only one school and we might surmise\
that there is some unique reason for this schools poor performance.
